```python
#HANDLING MISSING VALUES(using mean and scikit learn)

from sklearn.impute import SimpleImputer
si=SimpleImputer(missing_values=np.nan,strategy='mean')
ds[['Age', 'Salary']]=si.fit_transform(ds[['Age', 'Salary']])
```

```python
ds
```

```python
ds['Country'].unique()
```

```python
#ENCODING CATEGEORICAL COLUMN(onehotencoding)

from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(sparse_output=False)  # Corrected spelling
encoded_data = ohe.fit_transform(ds[['Country']])
encoded_data
```

```python
# CREATING COLUMNS
column_names = ohe.get_feature_names_out(['Country'])
encoded_df=pd.DataFrame(encoded_data, columns=column_names)
encoded_df
```

```python
#joinning original dataset and new dataframe

ds=pd.concat([ds,encoded_df],axis=1)
ds
```

```python
#drop actual categorical country column
ds=ds.drop(columns=['Country'])
ds
```

```python
#LABLE ENCODING

from sklearn.preprocessing import LabelEncoder
le= LabelEncoder()
ds['Purchased']=le.fit_transform(ds['Purchased'])
```

```python
ds
```

```python
#Feature scaling(reduse large data values to its 

from sklearn.preprocessing import StandardScaler
sc= StandardScaler()
ds[['Age', 'Salary']]=sc.fit_transform(ds[['Age', 'Salary']])
```

```python
#Divide dataset into independent (input)and dependant(output) variables

x=ds[['Country', 'Age', 'Salary', 'Purchased']]
y=ds[['Purchased']]
```

```python
#Splitting dataset as Training and testing

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x,y, test_size=0.2, random_state=1)
```
